```python
import os
import time
import threading
import json
import whisper
import tempfile
import base64
import glob
import logging
import librosa
from flask import Flask, render_template_string
from flask_socketio import SocketIO, emit
from pyngrok import ngrok
import soundfile as sf
from pydub import AudioSegment

# Configure logging
logging.basicConfig(level=logging.INFO)

# Configure ngrok
ngrok.set_auth_token("2s9UYBDAhKCGblwUTV8PPf5AhJg_75PDRoniQ2hhzvtmoy8SN")

# Flask app
app = Flask(__name__)
socketio = SocketIO(app, cors_allowed_origins="*")
whisper_model = whisper.load_model("base")  # Load Whisper model

# Translations dictionary
translations = {
    'fa': {
        'title': 'تبدیل گفتار به متن',
        'placeholder': 'متن تبدیل‌شده اینجا نمایش داده می‌شود...',
        'lang_options': {
            'fa': 'فارسی',
            'en': 'انگلیسی',
            'fr': 'فرانسوی',
            'de': 'آلمانی',
            'ar': 'عربی',
            'it': 'ایتالیایی',
            'ru': 'روسی',
            'es': 'اسپانیایی'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'در حال پردازش صدا...',
        'ready': 'متن آماده است',
        'error': 'خطا در پردازش: {}',
        'dir': 'rtl',
        'align': 'right'
    },
    'en': {
        'title': 'Speech to Text',
        'placeholder': 'Transcribed text will appear here...',
        'lang_options': {
            'fa': 'Persian',
            'en': 'English',
            'fr': 'French',
            'de': 'German',
            'ar': 'Arabic',
            'it': 'Italian',
            'ru': 'Russian',
            'es': 'Spanish'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Processing audio...',
        'ready': 'Text is ready',
        'error': 'Error in processing: {}',
        'dir': 'ltr',
        'align': 'left'
    },
    'fr': {
        'title': 'Parole à Texte',
        'placeholder': 'Le texte transcrit apparaîtra ici...',
        'lang_options': {
            'fa': 'Persan',
            'en': 'Anglais',
            'fr': 'Français',
            'de': 'Allemand',
            'ar': 'Arabe',
            'it': 'Italien',
            'ru': 'Russe',
            'es': 'Espagnol'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Traitement de l\'audio...',
        'ready': 'Texte prêt',
        'error': 'Erreur de traitement: {}',
        'dir': 'ltr',
        'align': 'left'
    },
    'de': {
        'title': 'Sprache zu Text',
        'placeholder': 'Der transkribierte Text erscheint hier...',
        'lang_options': {
            'fa': 'Persisch',
            'en': 'Englisch',
            'fr': 'Französisch',
            'de': 'Deutsch',
            'ar': 'Arabisch',
            'it': 'Italienisch',
            'ru': 'Russisch',
            'es': 'Spanisch'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Audio wird verarbeitet...',
        'ready': 'Text ist bereit',
        'error': 'Fehler bei der Verarbeitung: {}',
        'dir': 'ltr',
        'align': 'left'
    },
    'ar': {
        'title': 'تحويل الكلام إلى نص',
        'placeholder': 'سيظهر النص المحول هنا...',
        'lang_options': {
            'fa': 'فارسي',
            'en': 'إنجليزي',
            'fr': 'فرنسي',
            'de': 'ألماني',
            'ar': 'عربي',
            'it': 'إيطالي',
            'ru': 'روسي',
            'es': 'إسباني'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'جاري معالجة الصوت...',
        'ready': 'النص جاهز',
        'error': 'خطأ في المعالجة: {}',
        'dir': 'rtl',
        'align': 'right'
    },
    'it': {
        'title': 'Parlato a Testo',
        'placeholder': 'Il testo trascritto apparirà qui...',
        'lang_options': {
            'fa': 'Persiano',
            'en': 'Inglese',
            'fr': 'Francese',
            'de': 'Tedesco',
            'ar': 'Arabo',
            'it': 'Italiano',
            'ru': 'Russo',
            'es': 'Spagnolo'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Elaborazione audio...',
        'ready': 'Testo pronto',
        'error': 'Errore nell\'elaborazione: {}',
        'dir': 'ltr',
        'align': 'left'
    },
    'ru': {
        'title': 'Речь в Текст',
        'placeholder': 'Транскрибированный текст появится здесь...',
        'lang_options': {
            'fa': 'Персидский',
            'en': 'Английский',
            'fr': 'Французский',
            'de': 'Немецкий',
            'ar': 'Арабский',
            'it': 'Итальянский',
            'ru': 'Русский',
            'es': 'Испанский'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Обработка аудио...',
        'ready': 'Текст готов',
        'error': 'Ошибка обработки: {}',
        'dir': 'ltr',
        'align': 'left'
    },
    'es': {
        'title': 'Voz a Texto',
        'placeholder': 'El texto transcrito aparecerá aquí...',
        'lang_options': {
            'fa': 'Persa',
            'en': 'Inglés',
            'fr': 'Francés',
            'de': 'Alemán',
            'ar': 'Árabe',
            'it': 'Italiano',
            'ru': 'Ruso',
            'es': 'Español'
        },
        'start_button': '🎤 Start Recording',
        'stop_button': '■ Stop Recording',
        'loading': 'Procesando audio...',
        'ready': 'Texto listo',
        'error': 'Error en el procesamiento: {}',
        'dir': 'ltr',
        'align': 'left'
    }
}

# HTML template with JS for manual audio recording
template = """
<!doctype html>
<html lang="{{ lang }}">
  <head>
    <meta charset="utf-8">
    <title id="page-title">{{ trans['title'] }}</title>
    <script src="https://cdn.socket.io/4.4.1/socket.io.min.js"></script>
    <style>
      body {
        font-family: sans-serif;
        text-align: {{ trans['align'] }};
        padding: 20px;
        background: #f4f7f9;
        direction: {{ trans['dir'] }};
        transition: all 0.3s ease;
      }
      h1 {
        color: #2c3e50;
        font-weight: bold;
        animation: fadeIn 0.5s;
      }
      .box {
        background: white;
        padding: 25px;
        border-radius: 15px;
        width: 90%;
        max-width: 600px;
        margin: auto;
        box-shadow: 0 4px 20px rgba(0,0,0,0.1);
        transition: box-shadow 0.3s;
      }
      .box:hover { box-shadow: 0 6px 25px rgba(0,0,0,0.15); }
      select, button {
        padding: 12px;
        margin: 12px 5px;
        width: 48%;
        border-radius: 10px;
        border: 1px solid #bdc3c7;
        font-size: 16px;
        transition: border 0.3s;
      }
      select:focus, button:focus { border: 1px solid #3498db; }
      button {
        background: #27ae60;
        color: white;
        cursor: pointer;
        font-weight: bold;
        transition: background 0.3s, transform 0.2s;
      }
      button:hover { background: #219d54; transform: scale(1.02); }
      #stop-button { background: #e74c3c; }
      #stop-button:hover { background: #c0392b; }
      #result {
        font-size: 1.5em;
        color: #2c3e50;
        margin-top: 20px;
        min-height: 50px;
      }
      .error { color: #e74c3c; font-weight: bold; }
      .loading { color: #3498db; animation: pulse 1s infinite; }
      @keyframes pulse { 0% { opacity: 1; } 50% { opacity: 0.5; } 100% { opacity: 1; } }
      @keyframes fadeIn { from { opacity: 0; } to { opacity: 1; } }
      @media (max-width: 480px) {
        .box { padding: 15px; }
        h1 { font-size: 24px; }
        select, button { width: 100%; margin: 10px 0; }
      }
    </style>
    <script>
      const socket = io();
      const translations = {{ translations_json | safe }};
      let mediaRecorder = null;
      let isRecording = false;

      function updateUI(lang) {
        const trans = translations[lang] || translations['en'];
        document.getElementById('page-title').innerText = trans.title;
        document.querySelector('h1').innerText = '🎤 ' + trans.title;
        document.getElementById('result').innerText = trans.placeholder;
        document.getElementById('start-button').innerText = trans.start_button;
        document.getElementById('stop-button').innerText = trans.stop_button;

        const langSelect = document.getElementById('lang-select');
        for (let option of langSelect.options) {
          option.text = trans.lang_options[option.value];
        }

        document.body.style.direction = trans.dir;
        document.body.style.textAlign = trans.align;

        document.querySelectorAll('h1, select, button, #result').forEach(el => {
          el.style.animation = 'fadeIn 0.5s';
        });
      }

      socket.on('transcription', text => {
        document.getElementById('result').innerText = text;
        document.getElementById('result').className = '';
      });

      socket.on('error', error => {
        document.getElementById('result').innerText = translations[document.getElementById('lang-select').value].error.replace('{}', error);
        document.getElementById('result').className = 'error';
      });

      function showLoading() {
        document.getElementById('result').innerText = translations[document.getElementById('lang-select').value].loading;
        document.getElementById('result').className = 'loading';
      }

      function startRecording() {
        if (isRecording) return;
        navigator.mediaDevices.getUserMedia({ audio: true }).then(stream => {
          const mimeType = MediaRecorder.isTypeSupported('audio/wav') ? 'audio/wav' : 'audio/webm';
          console.log("Using mimeType:", mimeType);
          if (mimeType !== 'audio/wav') {
            alert('Your browser does not support WAV format. The file will be recorded as WebM and converted by the server.');
          }
          mediaRecorder = new MediaRecorder(stream, { mimeType: mimeType });
          const audioChunks = [];

          mediaRecorder.ondataavailable = event => {
            audioChunks.push(event.data);
          };

          mediaRecorder.onstop = () => {
            const audioBlob = new Blob(audioChunks, { type: mimeType });
            const reader = new FileReader();
            reader.onloadend = () => {
              socket.emit('audio', { data: reader.result, lang: document.getElementById('lang-select').value, mimeType: mimeType });
            };
            reader.readAsDataURL(audioBlob);
            isRecording = false;
            document.getElementById('start-button').disabled = false;
            document.getElementById('stop-button').disabled = true;
          };

          mediaRecorder.start();
          isRecording = true;
          document.getElementById('start-button').disabled = true;
          document.getElementById('stop-button').disabled = false;
          showLoading();
        }).catch(err => {
          document.getElementById('result').innerText = `Microphone error: ${err}`;
          document.getElementById('result').className = 'error';
        });
      }

      function stopRecording() {
        if (mediaRecorder && isRecording) {
          mediaRecorder.stop();
        }
      }

      window.onload = function() {
        updateUI('{{ lang }}');
        document.getElementById('lang-select').addEventListener('change', function() {
          updateUI(this.value);
        });
        document.getElementById('stop-button').disabled = true;
      };
    </script>
  </head>
  <body>
    <div class="box">
      <h1>🎤 {{ trans['title'] }}</h1>
      <select id="lang-select">
        {% for code, name in trans['lang_options'].items() %}
          <option value="{{ code }}" {% if code == lang %}selected{% endif %}>{{ name }}</option>
        {% endfor %}
      </select>
      <button onclick="startRecording()" id="start-button">{{ trans['start_button'] }}</button>
      <button onclick="stopRecording()" id="stop-button">{{ trans['stop_button'] }}</button>
      <p id="result" class="">{{ trans['placeholder'] }}</p>
    </div>
  </body>
</html>
"""

# Function to convert WebM to WAV
def convert_to_wav(input_path, output_path):
    try:
        audio = AudioSegment.from_file(input_path)
        audio.export(output_path, format="wav")
        logging.info(f"File {input_path} converted to {output_path}")
        return output_path
    except Exception as e:
        logging.error(f"Error converting file to WAV: {str(e)}")
        raise

# Function to preprocess audio
def preprocess_audio(audio_path):
    try:
        # Check if file exists
        logging.info(f"Checking file: {audio_path}")
        if not os.path.exists(audio_path):
            raise FileNotFoundError(f"File {audio_path} does not exist")

        # Load file with soundfile
        audio, sr = sf.read(audio_path)
        logging.info(f"File loaded with sample rate: {sr}")

        # If sample rate is different, resample to 16000
        if sr != 16000:
            logging.info(f"Resampling from {sr} to 16000")
            audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)
            sr = 16000

        # Save processed file
        with tempfile.NamedTemporaryFile(delete=False, suffix='.wav') as f:
            sf.write(f.name, audio, sr)
            logging.info(f"Processed file saved at: {f.name}")
            return f.name

    except Exception as e:
        logging.error(f"Error in audio preprocessing: {str(e)}")
        raise

# Function to clean up old files
def cleanup_old_files():
    while True:
        try:
            audio_files = glob.glob("static/audio/*.wav")
            current_time = time.time()
            for file in audio_files:
                if current_time - os.path.getmtime(file) > 3600:
                    os.remove(file)
                    logging.info(f"Deleted old file: {file}")
        except Exception as e:
            logging.error(f"Error in cleanup: {e}")
        time.sleep(3600)

# Main route
@app.route('/')
def index():
    lang = 'en'  # Default to English
    trans = translations.get(lang, translations['en'])
    translations_json = json.dumps(translations)
    return render_template_string(template, trans=trans, lang=lang, translations_json=translations_json)

# Handle audio data
@socketio.on('audio')
def handle_audio(data):
    try:
        audio_data = base64.b64decode(data['data'].split(',')[1])
        lang = data['lang']
        mime_type = data.get('mimeType', 'audio/wav')
        logging.info(f"Received file with mimeType: {mime_type}")

        # Save temporary file
        suffix = '.webm' if mime_type == 'audio/webm' else '.wav'
        with tempfile.NamedTemporaryFile(delete=False, suffix=suffix) as f:
            f.write(audio_data)
            f.flush()
            logging.info(f"Temporary audio file saved at: {f.name}")
            os.system(f"file {f.name}")  # Check file type

            # If file is WebM, convert to WAV
            input_path = f.name
            if mime_type == 'audio/webm':
                wav_path = input_path.replace('.webm', '.wav')
                convert_to_wav(input_path, wav_path)
                input_path = wav_path
            else:
                wav_path = input_path

            # Preprocess audio
            processed_path = preprocess_audio(input_path)
            # Transcribe speech to text
            result = whisper_model.transcribe(processed_path, language=lang if lang != 'auto' else None)
            os.remove(input_path)
            if input_path != wav_path:
                os.remove(wav_path)
            os.remove(processed_path)
            text = result['text']
            logging.info(f"Transcribed text: {text}")
            emit('transcription', text, broadcast=True)
    except Exception as e:
        error = translations.get(data['lang'], translations['en'])['error'].format(str(e))
        logging.error(error)
        emit('error', str(e))

# Run server
if __name__ == '__main__':
    try:
        os.makedirs("static/audio", exist_ok=True)
        http_tunnel = ngrok.connect(5000, bind_tls=True)
        public_url = http_tunnel.public_url
        print(f"Flask app running at: {public_url}")
        threading.Thread(target=cleanup_old_files, daemon=True).start()
        socketio.run(app, port=5000, debug=False)
    except Exception as e:
        logging.error(f"Error starting server: {e}")
```
